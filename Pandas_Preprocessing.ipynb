# <p style="background-color:midnightblue; font-family:calibri; font-size:130%; color:white; text-align:center; border-radius:10px 10px; padding:15px">Step 7: Correlation Analysis</p>

```python
corr = data.corr(method="spearman")      #Pearson     #Kendall 
fig, ax = plt.subplots(figsize=(10, 6))

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask,1)] = True

sns.heatmap(corr, cmap="Reds" , mask=mask ,annot=True, ax=ax)
```
